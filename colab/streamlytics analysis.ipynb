```python
# Clean 'rating' and 'duration' columns

# Standardize 'rating' values based on common rating systems
def clean_rating(rating):
    # Ensure rating is a string before checking for 'min' or 'Season'
    if 'min' in str(rating) or 'Season' in str(rating):
        return 'Unknown'  # These are duration values, not ratings
    return rating

# Apply the cleaning function after converting the column to string type to handle any non-string values
df_netflix_titles['rating'] = df_netflix_titles['rating'].astype(str).apply(clean_rating)

# Extract movie duration in minutes and TV show duration in seasons
# Ensure 'duration' column is string type before applying string operations
df_netflix_titles['movie_duration_minutes'] = df_netflix_titles[df_netflix_titles['type'] == 'Movie']['duration'].astype(str).str.extract('(\d+) min').astype(float)
df_netflix_titles['tv_show_seasons'] = df_netflix_titles[df_netflix_titles['type'] == 'TV Show']['duration'].astype(str).str.extract('(\d+) Season').astype(float)

# Fill NaN values (for entries that are not movies/TV shows or 'Unknown')
df_netflix_titles['movie_duration_minutes'].fillna(0, inplace=True)
df_netflix_titles['tv_show_seasons'].fillna(0, inplace=True)

# Convert to int, as durations are whole numbers
df_netflix_titles['movie_duration_minutes'] = df_netflix_titles['movie_duration_minutes'].astype(int)
df_netflix_titles['tv_show_seasons'] = df_netflix_titles['tv_show_seasons'].astype(int)

# Drop the original 'duration' column as it's been processed into new columns
df_netflix_titles.drop(columns=['duration'], inplace=True)

print("Cleaning of 'rating' and 'duration' columns complete.")
print('\nUnique values and their counts for cleaned Rating in df_netflix_titles:')
display(df_netflix_titles['rating'].value_counts())
print('\nFirst 5 rows of df_netflix_titles with new duration columns:')
display(df_netflix_titles.head())
```

```python
def preprocess_text(text):
    # Convert to string to handle potential non-string inputs (like floats for NaN)
    text = str(text)
    # Replace 'nan' string with empty string if it appears after str conversion of actual NaN
    if text == 'nan':
        return ''
    # Replace 'Unknown' with empty string as it was a fillna value previously
    if text == 'Unknown':
        return ''
    return text.lower()

# Apply the preprocessing function to the relevant columns
df_netflix_titles['listed_in'] = df_netflix_titles['listed_in'].apply(preprocess_text)
df_netflix_titles['director'] = df_netflix_titles['director'].apply(preprocess_text)
df_netflix_titles['cast'] = df_netflix_titles['cast'].apply(preprocess_text)
df_netflix_titles['description'] = df_netflix_titles['description'].apply(preprocess_text)

print('Textual features in df_netflix_titles have been preprocessed (NaNs handled, converted to lowercase).')
print('\nFirst 5 rows of preprocessed columns:')
display(df_netflix_titles[['listed_in', 'director', 'cast', 'description']].head())
```

```python
df_netflix_titles['combined_features'] = (
    df_netflix_titles['listed_in'] + ' ' +
    df_netflix_titles['director'] + ' ' +
    df_netflix_titles['cast'] + ' ' +
    df_netflix_titles['description']
)

print("Combined textual features into 'combined_features' column.")
print('\nFirst 5 rows of the new \'combined_features\' column:')
display(df_netflix_titles[['title', 'combined_features']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000) # Limiting features for efficiency

# Fit and transform the 'combined_features' column to create the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df_netflix_titles['combined_features'])

print('TF-IDF matrix created successfully.')
print(f'Shape of TF-IDF matrix: {tfidf_matrix.shape}')
print(f'Type of TF-IDF matrix: {type(tfidf_matrix)}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

print('Cosine similarity matrix created successfully.')
print(f'Shape of cosine similarity matrix: {cosine_sim.shape}')
```

```python
indices = pd.Series(df_netflix_titles.index, index=df_netflix_titles['title']).drop_duplicates()

print('Indices Series created successfully.')
print('First 5 entries of the indices Series:')
display(indices.head())
```

```python
def get_recommendations(title, cosine_sim=cosine_sim, df=df_netflix_titles, indices=indices):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies (excluding itself)
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title'].iloc[movie_indices]

print("Recommendation function 'get_recommendations' defined.")

# Test the function with an example
print("\nRecommendations for 'Dick Johnson Is Dead':")
display(get_recommendations('Dick Johnson Is Dead'))
```

```python
print("\nRecommendations for 'Kota Factory':")
display(get_recommendations('Kota Factory'))

print("\nRecommendations for 'Blood & Water':")
display(get_recommendations('Blood & Water'))
```
